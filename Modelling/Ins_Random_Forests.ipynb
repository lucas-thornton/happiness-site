```python
all_data_developed = all_data[all_data['Country'].isin(Developed_countries)]
all_data_developing = all_data[all_data['Country'].isin(Developing_countries)]
all_data_underdeveloped = all_data[all_data['Country'].isin(Underdeveloped_countries)]
```

```python
happiness_developed = all_data_developed[['Happiness']]
happiness_developing = all_data_developing[['Happiness']]
happiness_underdeveloped = all_data_underdeveloped[['Happiness']]
```

```python
data_developed = all_data_developed.drop(['Country', 'Happiness'], axis=1)
data_developing = all_data_developing.drop(['Country', 'Happiness'], axis=1)
data_underdeveloped = all_data_underdeveloped.drop(['Country', 'Happiness'], axis=1)
```

```python
full_data_developed = data_developed.fillna(data_developed.mean())
full_data_developing = data_developing.fillna(data_developing.mean())
full_data_underdeveloped = data_underdeveloped.fillna(data_underdeveloped.mean())
full_data_developed
```

```python
# from sklearn.preprocessing import OneHotEncoder
# from keras.utils import to_categorical
# # enc = OneHotEncoder(handle_unknown='ignore')
# encoded = to_categorical(full_data_developed)
# encoded

onehot_developed_reg = pd.get_dummies(full_data_developed.Region, prefix='Region')
onehot_developed_rel = pd.get_dummies(full_data_developed.Religion, prefix='Religion')
onehot_developed_lan = pd.get_dummies(full_data_developed.Languages, prefix='Language')
drop_developed = full_data_developed.drop(['Region', 'Languages', 'Religion'], axis=1)
total_developed = drop_developed.join(onehot_developed_lan)
total_developed = total_developed.join(onehot_developed_rel)
total_developed = total_developed.join(onehot_developed_reg)
```

```python
onehot_developing_reg = pd.get_dummies(full_data_developing.Region, prefix='Region')
onehot_developing_rel = pd.get_dummies(full_data_developing.Religion, prefix='Religion')
onehot_developing_lan = pd.get_dummies(full_data_developing.Languages, prefix='Language')
drop_developing = full_data_developing.drop(['Region', 'Languages', 'Religion'], axis=1)
total_developing = drop_developing.join(onehot_developing_lan)
total_developing = total_developing.join(onehot_developing_rel)
total_developing = total_developing.join(onehot_developing_reg)
```

```python
onehot_underdeveloped_reg = pd.get_dummies(full_data_underdeveloped.Region, prefix='Region')
onehot_underdeveloped_rel = pd.get_dummies(full_data_underdeveloped.Religion, prefix='Religion')
onehot_underdeveloped_lan = pd.get_dummies(full_data_underdeveloped.Languages, prefix='Language')
drop_underdeveloped = full_data_underdeveloped.drop(['Region', 'Languages', 'Religion', 'Antidepressants (per cap)'], axis=1)
total_underdeveloped = drop_underdeveloped.join(onehot_underdeveloped_lan)
total_underdeveloped = total_underdeveloped.join(onehot_underdeveloped_rel)
total_underdeveloped = total_underdeveloped.join(onehot_underdeveloped_reg)
```

```python
total_underdeveloped.to_csv("allfactorsunder.csv")
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler() 
developed_scaled = scaler.fit_transform(total_developed) 
developing_scaled = scaler.fit_transform(total_developing) 
underdeveloped_scaled = scaler.fit_transform(total_underdeveloped) 
happ_developed_scaled = scaler.fit_transform(happiness_developed)
happ_developing_scaled = scaler.fit_transform(happiness_developing)
happ_underdeveloped_scaled = scaler.fit_transform(happiness_underdeveloped)
```

```python
# Create a random forest classifier
rf = RandomForestRegressor(n_estimators=200)
rf = rf.fit(developed_scaled, happ_developed_scaled)
rf.score(developed_scaled, happ_developed_scaled)
```

```python
# Random Forests in sklearn will automatically calculate feature importance
importances = rf.feature_importances_
```

```python
# We can sort the features by their importance
sorted(zip(rf.feature_importances_, total_developed), reverse=True)
```
